```python
import pandas as pd
```

```python
alcohol_df = pd.read_csv("STNAME x ABODALC.csv")
```

```python
alcohol_df.head()
```

```python
alcohol_df2 = alcohol_df.set_index("STATE NAME").sort_index()
alcohol_df3 = alcohol_df2.loc[alcohol_df2['RC-ALCOHOL DEPENDENCE OR ABUSE - PAST YEAR'].isin(['1 - Yes', '0 - No/Unknown'])]
alcohol_df3 = alcohol_df3[['RC-ALCOHOL DEPENDENCE OR ABUSE - PAST YEAR', "Count SE"]]
alcohol_df_yes = alcohol_df3.loc[alcohol_df3['RC-ALCOHOL DEPENDENCE OR ABUSE - PAST YEAR'] == '1 - Yes']
alcohol_df_no = alcohol_df3.loc[alcohol_df3['RC-ALCOHOL DEPENDENCE OR ABUSE - PAST YEAR'] != '1 - Yes']
alcohol_df_yes.rename(columns = {'Count SE': 'Yes'}, inplace = True)
alcohol_df_yes.drop("RC-ALCOHOL DEPENDENCE OR ABUSE - PAST YEAR",axis = 1,  inplace = True)
alcohol_df_no.rename(columns = {'Count SE': 'No'}, inplace = True)
alcohol_df_no.drop("RC-ALCOHOL DEPENDENCE OR ABUSE - PAST YEAR",axis = 1,  inplace = True)
final_df = alcohol_df_yes
final_df['No'] = alcohol_df_no["No"]
```

```python
final_df['Total'] =  final_df.apply(lambda x: x['Yes'] + x['No'], axis=1)
final_df.reset_index(inplace = True)
```

```python
final_df.head()
```

```python
final_df['Yes%'] = final_df.apply(lambda x: x['Yes']/x['Total']*100, axis = 1)
final_df['No%'] = final_df.apply(lambda x: x['No']/x['Total']*100, axis = 1)
```

```python
final_df.head()
```

```python
final_df = final_df[~final_df['STATE NAME'].isin(['DISTRICT OF COLUMBIA', 'Overall'])]
```

```python
final_df.to_csv('alcohol_data.csv', index=False)
```
