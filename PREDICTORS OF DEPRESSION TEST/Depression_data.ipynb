```python
import pandas as pd
```

```python
depression_df = pd.read_csv('STNAME x ADWRDEPR.csv')
```

```python
depression_df2 = depression_df.set_index("STATE NAME").sort_index()
```

```python
depression_df2
```

```python
depression_df3 = depression_df2.loc[depression_df2['WHEN PRBLMS WORST FELT SAD/EMPTY/DPRSD EVERY DAY'].isin(['1 - Yes', '2 - No'])]
                                                                                                            
                                                                                                            
```

```python
depression_df3 = depression_df3[['WHEN PRBLMS WORST FELT SAD/EMPTY/DPRSD EVERY DAY', "Count SE"]]
```

```python
depression_df3.head()
```

```python
depression_df_yes = depression_df3.loc[depression_df3['WHEN PRBLMS WORST FELT SAD/EMPTY/DPRSD EVERY DAY'] == '1 - Yes']

```

```python
depression_df_yes.rename(columns = {'Count SE': 'Yes'}, inplace = True)
depression_df_yes.drop("WHEN PRBLMS WORST FELT SAD/EMPTY/DPRSD EVERY DAY",axis = 1,  inplace = True)
depression_df_yes.head()
```

```python
depression_df_no = depression_df3.loc[depression_df3['WHEN PRBLMS WORST FELT SAD/EMPTY/DPRSD EVERY DAY'] != '1 - Yes']
```

```python
depression_df_no.rename(columns = {'Count SE': 'No'}, inplace = True)
depression_df_no.drop("WHEN PRBLMS WORST FELT SAD/EMPTY/DPRSD EVERY DAY",axis = 1,  inplace = True)
depression_df_no.head()
```

```python
final_df = depression_df_yes
final_df['No'] = depression_df_no["No"]
```

```python
final_df.head()
```

```python
final_df['Total'] =  final_df.apply(lambda x: x['Yes'] + x['No'], axis=1)
final_df.reset_index(inplace = True)
```

```python
final_df.head()
```

```python
final_df['Yes%'] = final_df.apply(lambda x: x['Yes']/x['Total']*100, axis = 1)
final_df['No%'] = final_df.apply(lambda x: x['No']/x['Total']*100, axis = 1)
```

```python
final_df.head()
```

```python
final_df = final_df[~final_df['STATE NAME'].isin(['DISTRICT OF COLUMBIA', 'Overall'])]
```

```python
final_df
```

```python
final_df.to_csv('depression_data.csv', index=False)
```
